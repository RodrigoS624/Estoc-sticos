```python
import random                   
from collections import Counter  #sirve para contar las frecuencias de estados
```

## Función de Simulación

```python
def cadenaMarkov(a, P, n): 
    #alpha: distribución inicial 
    # P: matriz de transición 
    #n: número de pasos

    num_estados = len(a)

    #Creamos una lista con los estados que hay
    estados = list(range(num_estados))

    #Creamos el estado inicial con base a alpha
    #(En realidad podríamos verlo como que estamos fingiendo dónde empezó nuestro evento con base a las probabilidades)
    X0 = random.choices(estados, weights=a)[0]

    #Vamos guardando los estados que se van generando
    trayectoria = [X0]

    for i in range(1, n + 1):

        #Agarramos el último estado generado 
        estado_actual = trayectoria[-1]

        #Escogemos la j-esima fila de la matriz P
        #Aquí lo que hacemos es determinar las probabilidades del siguiente paso
        probabilidades = P[estado_actual]

        # generamos el siguiente estado usando esas probabilidades
        siguiente_estado = random.choices(estados, weights=probabilidades)[0]

        #Guardamos el estado para ir construyendo la sución de eventos (Como estaba en el pizarron X_0 , X_1 , ...)
        trayectoria.append(siguiente_estado)

    return trayectoria
```

## Ejemplificando con el Tablero de Colores

```python
#Estados del sistema
    # 0 = Naranja
    # 1 = Azul
    # 2 = Verde
    # 3 = Rojo

#Probabilidades de cada evento (en estee caso es uniforme)
a = [0.25, 0.25, 0.25, 0.25]

#Matriz de transición
P = [

    [1/4, 0,  1/4, 1/2],  

    [1/2, 1/4, 0,  1/4],  

    [1/4, 1/2, 1/4, 0],    

    [0,  1/4, 1/2, 1/4]  
    
    ]

#Pasos
n = 100000


trayectoria = cadenaMarkov(a, P, n)
```

Ahora hagamos un pequeño analisis de datosde los resultados obtenidos en la simulación:

```python
#Contamos cuántas veces aparece cada estado
conteo = Counter(trayectoria)

print("Frecuencia relativa de cada estado:\n")

for estado in range(4):

    frecuencia = conteo[estado] / len(trayectoria)

    print("Estado", estado, ":", frecuencia)
```

Podemos notar que en efecto, los estados se estabilizan y en este caso son equiprobables dado su probabilidad
